```python
import numpy as np

import matplotlib.pyplot as plt
import networkx as nx
```

```python
graph = nx.read_adjlist("g5.txt", nodetype=int)
    
```

```python
plt.axes().set_aspect('equal', adjustable='datalim')
nx.draw_shell(graph, with_labels=True, node_color = "white", font_weight='bold', font_color='black')
plt.savefig('G1.png')
```

```python
plt.axes().set_aspect('equal', adjustable='datalim')
pos = {1: (0,0), 2:(1,0), 3: (0,1), 4: (1,1), 5: (0.5 ,2.0), 6:(2,0), 7: (3,0), 8: (2,1), 9: (3,1), 10: (2.5 , 2.0), 11: (4,0), 12: (5,0) }
nx.draw(graph, pos = pos, with_labels=True, node_color = "white",  font_weight='bold', font_color='black')
plt.savefig('G2.png')
```

```python
i = 1
for c in nx.connected_components(graph):
    
    print("For connected component ", i, ":")
    subg = graph.subgraph(c)
    print("Number of nodes: ", subg.number_of_nodes())
    print("Number of edges: ", subg.number_of_edges())
    print("Diameter is: ", nx.diameter(subg))
    print("Radius is: ", nx.radius(subg))
    for v in subg.nodes:
        print("Excentricitet for node ", v, "is: ", nx.eccentricity(subg,v))
        print("Degree of node ", v, "is: ", subg.degree(v))
    print()
    i+=1
```

```python

for c in nx.connected_components(graph):
    subg = graph.subgraph(c)
    for  u in subg.nodes:
        for v in subg.nodes:
            n = nx.shortest_path(subg, u, v)
            l = nx.shortest_path_length(subg, u, v)
            if l == nx.diameter(subg):
                m = tuple(n)
                if len(m) == 2:
                    diam = m
    helper = nx.Graph()
    helper.add_edges_from([diam])
    for u, v in subg.edges:
        if subg.has_edge(u, v) and helper.has_edge(u, v):
            nx.set_edge_attributes(graph, name='color', values={(u, v): 'red'})

        else:
            nx.set_edge_attributes(graph, name='color', values={(u, v): 'black'})
colorsDiam = [graph[u][v]['color'] for u, v in graph.edges]   
plt.axes().set_aspect('equal', adjustable='datalim')
nx.draw(graph, pos = pos, edge_color=colorsDiam, with_labels=True, node_color = "white", font_weight='bold', font_color='black')
    
```
